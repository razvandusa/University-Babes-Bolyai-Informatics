# 2. Se dau mai multe imagini (salvate in folder-ul "data/images"). Se cere:
    - sa se vizualizeze una din imagini
    - daca imaginile nu aceeasi dimensiune, sa se redimensioneze toate la 128 x 128 pixeli si sa se vizualizeze imaginile intr-un cadru tabelar.
    - sa se transforme imaginile in format gray-levels si sa se vizualizeze
    - sa se blureze o imagine si sa se afiseze in format "before-after"
    - sa se identifice muchiile intr-o imagine si sa se afiseze in format "before-after"

```python
import numpy as np
import PIL
import os
import matplotlib.pyplot as plt
from PIL import ImageFilter, Image

# Path for the images folder
images_folder_path = 'data/images'
images = []

for file in os.listdir(images_folder_path):
    path = os.path.join(images_folder_path, file)
    image = Image.open(path)
    images.append(image)
```

## 2.1 Sa se vizualizeze una din imagini

```python
# 2.1 Sa se vizualizeze una din imagini
def showFirstImage():
    # Opening and displaying the image
    images[0].show()
showFirstImage()
```

## 2.2 Daca imaginile nu aceeasi dimensiune, sa se redimensioneze toate la 128 x 128 pixeli si sa se vizualizeze imaginile intr-un cadru tabelar.

```python
def resizeAndShowImages():
    resized_images = []
    for image in images:
        resized_img = image.resize((128, 128))
        resized_images.append(resized_img)
    # Calculate grid dimensions
    num_images = len(resized_images)
    # Dimensiune grid
    cols = int(4)
    rows = int(num_images // cols + 1)
    plt.figure(figsize=(10,10))
    # Afisare imagini
    for i, img in enumerate(resized_images):
        plt.subplot(rows, cols, i + 1)
        plt.imshow(img)
        plt.axis('off')
    plt.show()
resizeAndShowImages()
```

## 2.3 Sa se transforme imaginile in format gray-levels si sa se vizualizeze

```python
def showGrayLevelImages():
    for image in images:
        gray = image.convert('L') # Transformare in grayscale
        plt.imshow(gray, cmap='gray')
        plt.axis('off')
        plt.show()
showGrayLevelImages()
```

## 2.4 Sa se blureze o imagine si sa se afiseze in format "before-after"

```python
def showBlurredImage():
    image = images[0]
    blurred = image.filter(ImageFilter.BLUR)
    plt.figure(figsize=(8,4))

    plt.subplot(1,2,1)
    plt.imshow(image)
    plt.title('Before')
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.imshow(blurred)
    plt.title('After')
    plt.axis('off')

    plt.show()
showBlurredImage()
```

## 2.5 Sa se identifice muchiile intr-o imagine si sa se afiseze in format "before-after"

```python
def showEdges():
    image = images[0]
    edges = image.filter(ImageFilter.FIND_EDGES)
    plt.figure(figsize=(8,4))

    plt.subplot(1,2,1)
    plt.imshow(image)
    plt.title('Original')
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.imshow(edges)
    plt.title('Edges')
    plt.axis('off')

    plt.show()
showEdges()
```
